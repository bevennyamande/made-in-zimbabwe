```python
import csv
import json
import requests
import pandas as pd

query = 'zimbabwe'
URL = f'https://api.github.com/search/users?q={query}'

try:
    response = requests.get(URL)
    data = response.json()
    
    with open('results.json', 'r+') as fhandle:
        json.dump(data, fhandle)
        
except Exception as e:
    print(e)
df = pd.read_json(r'results.json')
users = {}

# login, score, avatar_url, url
for item in df['items']:
#     print(item['login'],item['url'],item['score'],item['avatar_url'])
    users.update({ item['login']: [item['url'], item['score'], item['avatar_url']] })
users


    
```
